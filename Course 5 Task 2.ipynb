```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns 
%matplotlib inline
```

```python
credit2.groupby('age_bucket')['age_bucket'].count()
```

```python
fg = sns.catplot('age_bucket', data=credit2, kind='count', aspect=1.5)
fg.set_xlabels('Age')
```

```python
# Number of men and women in each age range
credit2.groupby(['SEX', 'age_bucket'])['SEX'].count()
```

```python
g = sns.catplot('age_bucket', data=credit2, hue='SEX', kind='count', aspect=1.75)
g.set_xlabels('age_bucket')
```

```python
credit2.groupby('default payment next month')['default payment next month'].count()
```

```python
g = sns.catplot('age_bucket', data=credit2, hue='default payment next month', kind='count', aspect=1.75)
g.set_xlabels('age_bucket')
```

```python
g = sns.catplot('SEX', data=credit2, hue='default payment next month', kind='count', aspect=1.75)
g.set_xlabels('SEX')
```

```python
g = sns.catplot('limit_bal_bucket', data=credit2, hue='default payment next month', kind='count', aspect=5.25)
g.set_xlabels('limit_bal_bucket')
```

```python
g = sns.catplot('PAY_0', data=credit2, hue='default payment next month', kind='count', aspect=1.75)
g.set_xlabels('PAY_0')
```

```python
fig = sns.FacetGrid(credit2, hue='default payment next month', aspect=4)
fig.map(sns.kdeplot, 'AGE', shade=True)
oldest = credit2['AGE'].max()
fig.set(xlim=(0,oldest))
fig.set(title='Distribution of Age Grouped Default Prediction')
fig.add_legend()
```

```python
fig = sns.FacetGrid(credit2, hue='default payment next month', aspect=4)
fig.map(sns.kdeplot, 'PAY_0', shade=True)
oldest = credit2['PAY_0'].max()
fig.set(xlim=(-2,oldest))
fig.set(title='Distribution of PAY_0 Status Grouped Default Prediction')
fig.add_legend()
```

```python
fig = sns.FacetGrid(credit2, hue='default payment next month', aspect=4)
fig.map(sns.kdeplot, 'EDUCATION', shade=True)
oldest = credit2['EDUCATION'].max()
fig.set(xlim=(0,oldest))
fig.set(title='Distribution of Education Level Grouped Default Prediction')
fig.add_legend()
```
